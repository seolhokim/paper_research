```python
texts = ' '.join(raw_data.iloc[:10000,:]['comment_text'].values)
```

```python
p = re.compile(r"[\(\[].*?[\)\]]")
pp = re.compile("[^a-zA-Z ]")
```

```python
texts = texts.replace("  "," ").replace("  "," ").replace("  "," ")
```
